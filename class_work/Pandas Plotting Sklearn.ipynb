# Intro to Pandas

The most common moduel that we will be using is one called pandas. It allows us to use matrix like objects for data manipulation and analysis. This allows us to do things that will approximate what you do in stata which essentially expects a dataframe in memory

```python
import pandas as pd # this alias is ubiquitous but not necessary
from sklearn.linear_model import LinearRegression
import seaborn # graphing functions
```

```python
print(pd.DataFrame([[1,2],[3,4]]), '\n') # you can coerce other data into a dataframe to use it as a matrix
print(pd.DataFrame({'a':[1,3],'b':[2,4]}))
```

```python
df = pd.read_csv('IRIS.csv') # you can read in files with raw data that you would like to manipulate or analyze
```

```python
df.head(10) # head displayed the top n rows, default is 5
```

```python
df.tail() # tail shows the last rows
```

```python
df.describe() # describe gives you summary stats on the different variables
```

```python
df.sum() # functions are set up to run element wise, they go over colums or values within columns
```

```python
df.groupby('species').describe() # groupby allows to collect data with values in common and perform operations by group
```

```python
seaborn.lmplot('sepal_length','petal_length', data=df, col='species')
```
